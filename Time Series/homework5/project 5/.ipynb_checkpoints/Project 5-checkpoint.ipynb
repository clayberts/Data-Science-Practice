```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from matplotlib.dates import DateFormatter
%matplotlib inline
```

```python
df = pd.read_csv(r'./data/EuStockMarkets.csv', 
                 parse_dates=['Month'], 
                 index_col=['Month'])
df.head()
```

```python
fig, ax = plt.subplots()
plt.xlabel('Date')
plt.ylabel('CAC Value')
plt.title('CAC Value at End of Month')
plt.plot(df, color='black')
ax.xaxis.set_major_formatter(DateFormatter('%Y'))
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [7, 8]
sd = seasonal_decompose(df, 
                        period=12, 
                        model='additive').plot()
```

```python
sd = seasonal_decompose(df, 
                        period=12, 
                        model='multiplicative').plot()
```
