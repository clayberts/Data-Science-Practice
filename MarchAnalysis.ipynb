```python
data = pd.read_csv('/content/MarchMadness.csv')
```

```python
data
```

```python
x_data = data.iloc[:, 1:]
y_data = data.iloc[:, 0]
```

```python
sse = [] #SUM OF SQUARED ERROR
for k in range(1,11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(x_data)
    sse.append(km.inertia_)
```

```python
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11), y=sse)

g.set(xlabel ="Number of cluster (k)",
      ylabel = "Sum Squared Error",
      title ='Elbow Method')

plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 2)
kmeans.fit(x_data)
```

```python
kmeans.cluster_centers_
```

```python
pred = kmeans.fit_predict(x_data)
pred
```

```python
counts = {0:0, 1:0, 2:0, 3:0, 4:0}
for item in pred:
  if item == 0:
    counts[0] += 1
  elif item == 1:
    counts[1] += 1
  elif item == 2:
    counts[2] += 1
  elif item == 3:
    counts[3] += 1
  elif item == 4:
    counts[4] += 1

x_s = []
y_s = []
for key, val in counts.items():
  x_s.append(key)
  y_s.append(val)

print(x_s)
print(y_s)
```

```python
plt.figure(figsize=(12,5))
plt.bar(x_s, y_s)
plt.grid(True)
```
