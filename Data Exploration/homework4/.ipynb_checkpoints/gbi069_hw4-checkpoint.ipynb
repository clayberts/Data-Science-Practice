```python
#%% Q2 setup, 60pts
import scipy.stats as stats
import numpy.random as rn
#%% Q2a, 15pts
rn.seed(0)
N = 16;
m = 10**5;
# 10**5 by 16 = 16 + 16 + ... + 16
# use the uniform or binomial functions to generate the data
nHeads = (rn.uniform(size=(m, N))>0.5).sum(1)

#%%plot the histgram
# Your code is here
# suggest one line code

count, bins, bars = plt.hist(nHeads, bins=range(18))

plt.xlabel('nHeads')
plt.ylabel('Number of coins')
plt.title('Q2a: Histogram')
plt.show()

#%% Q2b, 10pts
# using plt.hist with parameter bins and density
#Your code is here

plt.hist(nHeads, bins=range(18), density=True)

plt.xlabel('k')
plt.ylabel('P(nHeads = k)')
plt.title('Q2b: Probability Mass Function')
plt.show()

#%% Q2c, 15pts

#Use the values of counts returned from a or b. 
#Use the function cumsum to get the CDF.
# Do not use scipy.stats.binom
# Your code is here

# print(np.cumsum(nHeads))
pdf = count / sum(count)
cdf = np.cumsum(pdf)
plt.plot(bins[1:], cdf)

plt.xlabel('k')
plt.ylabel('P(nHeads <= k)')
plt.title('Q2c: Cumulative Distribution Function')
plt.show()


#%% Q2d, 20pts
# Use the binomial distribution CDF (use scipy.stats.binom.cdf)

# Your code is here
# print(cdf)
theo_cdf = sci.binom.cdf(range(17), 16, 0.5)
# scatter plot
# Your code is here
plt.scatter(cdf, theo_cdf)

plt.xlabel('Empirical CDF')
plt.ylabel('Theoretical CDF')
plt.title('Q2d: scatter plot')
plt.show()

# line plot
#Your code is here
plt.plot(cdf, label="Empirical CDF", color='blue', marker='o')
plt.plot(theo_cdf, color='orange', marker='D', linestyle='dotted', label="Theoretical CDF")

plt.xlabel('k')
plt.ylabel('CDF')
plt.title('Q2d: line plot')
plt.legend()
plt.show()
plt.show()

# Loglog scale plot plt.loglog
#your code is here
plt.loglog(cdf, theo_cdf, 'o')

plt.xlim(10**-5, 1)
plt.xlabel('Empirical CDF')
plt.ylabel('Theoretical CDF')
plt.title('Q2d: loglog plot')
plt.show()
```

```python
##Q3, 20pts
#%% Q3a, 15pts
# the plot is not required
#Your code is here
x_data = np.array([2005, 2006, 2007, 2008, 2009])
y_data = np.array([12, 19, 29, 37, 45])
# print(x_data, y_data)
plt.scatter(x_data, y_data)

reg = linear_model.LinearRegression()
reg.fit(x_data.reshape(-1, 1), y_data.reshape(-1, 1))

print("y = " + str(reg.coef_[0, 0]) + "x " + str(reg.intercept_[0]))

plt.plot(x_data.reshape(-1, 1), reg.predict(x_data.reshape(-1, 1)), 'r')
plt.show()

#%%Q3b, 5pts
#Your code is here
# print(reg.predict([[2012]])[0][0])

#Output the result by print('Estimated sales in 2012 is ', XXX)
print('Estimated sales in 2012 is', reg.predict([[2012]])[0][0])
```
